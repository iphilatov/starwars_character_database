```python
def get_sources(alinksoup):
    media_list = []
    try:
        slct = alinksoup.find(string="Appearances").find_next().get_text()
        for line in slct.splitlines():
            if line != '':
                source = line.strip().strip(',')
                media_list.append(source)
                if source not in all_sources:
                    all_sources.append(source)
                else:
                    continue
        characters_media = ', '.join(media_list)
        return characters_media
    except AttributeError:
        return 'no sources'
        pass
        
            
def get_name(alinksoup):
    characters_name = alinksoup("h3")[0].get_text().strip()
    return characters_name

def get_species(alinksoup, lin):
    try:
        fv = alinksoup.find(string="species").find_next().get_text()
        for line in fv.splitlines():
            if line != '':
                vid = line.strip().strip(',')
                if vid not in all_species:
                    all_species.append(vid)
                else:
                    continue
        return(vid)
    except AttributeError:
        if alinksoup.find('div', class_='heading', string="droid") != None:
            return 'Droid'
        else:
            return 'unknown'
        
def get_gender(alinksoup):
    try:
        sx = alinksoup.find(string='Gender').find_next().get_text()
        return sx.strip().strip(',')
    except AttributeError:
        return 'no gender'
```

```python
def pronouns_define(rasa, pol):
    if rasa == 'Droid':
        mestoimenia = ['it', 'its', 'its']
    elif pol == 'no gender' and rasa != 'Droid':
        mestoimenia = ['they', 'them', 'their']
    else:
        if pol == 'Female':
            mestoimenia = ['she', 'her', 'her']
        elif pol == 'Male':
            mestoimenia = ['he', 'him', 'his']
    return mestoimenia

def race_define(imya):
    cur.execute('''SELECT species FROM starwars.species s 
    JOIN starwars.characters c ON c.id_species=s.idspecies 
    WHERE LOWER(`names`) = '{}';'''.format(imya.lower()))
    fetching = cur.fetchall()
    try:
        rasa = fetching[0][0]
    except:
        rasa = fetching[0]['species']
    return rasa

def appearance_define(imya):
    cur.execute('''SELECT source_name FROM starwars.characters c 
        JOIN starwars.appearances a ON c.idcharacters=a.id_characters 
        JOIN starwars.media m ON m.idmedia=a.id_media 
        WHERE LOWER(`names`) = '{}';'''.format(imya.lower()))
    fetching = cur.fetchall()
    try:
        list_of_appearances = [fe['source_name'] for fe in fetching]
    except TypeError:
        list_of_appearances = [fe[0] for fe in fetching]
    return list_of_appearances
    
```

```python
def charactersinmedia(ex):
    answer4a = ((input('What media? ')))
    if ex == '1':
        cur.execute('''SELECT `names` FROM starwars.characters c 
                JOIN starwars.appearances a ON c.idcharacters=a.id_characters 
                JOIN starwars.media m ON m.idmedia=a.id_media 
                WHERE LOWER(`source_name`) = '{}';'''.format(answer4a.lower()))
        fetchm = cur.fetchall()
        try:
            media_characters = [fm['names'] for fm in fetchm]
        except TypeError:
            media_characters = [fm[0] for fm in fetchm]
        print('Here is a list of all characters, who appeared in {}:'.format(answer4a))
        print(*media_characters, sep = ', ')
    if ex == '2':
        cur.execute('''SELECT count(*) as `number of characters` 
            FROM starwars.characters c 
            JOIN starwars.appearances a ON c.idcharacters=a.id_characters 
            JOIN starwars.media m ON m.idmedia=a.id_media 
            WHERE LOWER(`source_name`) = '{}';'''.format(answer4a.lower()))
        fetchm = cur.fetchall()
        try:
            O = fetchm[0]['number of characters']
        except TypeError:
            O = fetchm[0][0]
        print('The number of characters, who appeared in {} is'.format(answer4a), O)
```

```python
def charactersofspecies(ex):
    answer4a = ((input('What species? ')))
    if ex == '1':
        cur.execute('''SELECT `names` FROM starwars.characters c 
            JOIN starwars.species s ON s.idspecies=c.id_species 
            WHERE LOWER(`species`) = '{}';'''.format(answer4a.lower()))
        fetchm = cur.fetchall()
        try:
            species_characters = [fm['names'] for fm in fetchm]
        except TypeError:
            species_characters = [fm[0] for fm in fetchm]
        print('Here is a list of all characters, whose species is {}:'.format(answer4a))
        print(*media_characters, sep = ', ')
    if ex == '2':
        cur.execute('''SELECT count(*) as `number of characters` 
            FROM starwars.characters c 
            JOIN starwars.species s ON s.idspecies=c.id_species
            WHERE LOWER(`species`) = '{}'
            GROUP BY 'species';'''.format(answer4a.lower()))
        fetchm = cur.fetchall()
        try:
            O = fetchm[0]['number of characters']
        except TypeError:
            O = fetchm[0][0]
        print('The number of characters of {} species is'.format(answer4a), O)
```

```python
def mediawithspecies(ex):
    answer4a = ((input('What species? ')))
    if ex == '1':
        cur.execute('''SELECT DISTINCT(source_name) AS 'medias' FROM starwars.media m 
        JOIN starwars.appearances a ON m.idmedia=a.id_media
        JOIN starwars.characters c ON c.idcharacters=a.id_characters
        JOIN starwars.species s ON c.id_species=s.idspecies
        WHERE LOWER(`species`) = '{}';'''.format(answer4a.lower()))
        fetchm = cur.fetchall()
        try:
            species_media = [fm['medias'] for fm in fetchm]
        except TypeError:
            species_media = [fm[0] for fm in fetchm]
        print('Here is a list of all media, where {} individuals appear:'.format(answer4a))
        print(*species_media, sep = ', ')
    elif ex == '2':
        cur.execute('''SELECT count(DISTINCT(source_name)) AS 'medias' FROM starwars.media m 
        JOIN starwars.appearances a ON m.idmedia=a.id_media
        JOIN starwars.characters c ON c.idcharacters=a.id_characters
        JOIN starwars.species s ON c.id_species=s.idspecies
        WHERE LOWER(`species`) = '{}';'''.format(answer4a.lower()))
        fetchm = cur.fetchall()
        try:
            O = fetchm[0]['medias']
        except TypeError:
            O = fetchm[0][0]
        print('The number of media where {} individuals appeared is'.format(answer4a), O)
```

```python
def speciesinmedia(ex):
    answer4a = ((input('What media? ')))
    if ex == '1':
        cur.execute('''SELECT DISTINCT(`species`) FROM starwars.species s 
        JOIN starwars.characters c ON c.id_species=s.idspecies
        JOIN starwars.appearances a ON c.idcharacters=a.id_characters
        JOIN starwars.media m ON m.idmedia=a.id_media
        WHERE LOWER(`source_name`) = '{}';'''.format(answer4a.lower()))
        fetchm = cur.fetchall()
        try:
            media_species = [fm['species'] for fm in fetchm]
        except TypeError:
            media_species = [fm[0] for fm in fetchm]
        print('Here is a list of all species which appear in {}:'.format(answer4a))
        print(*media_species, sep = ', ')
    elif ex == '2':
        cur.execute('''SELECT COUNT(DISTINCT(`species`)) AS SDC FROM starwars.species s
        JOIN starwars.characters c ON c.id_species=s.idspecies
        JOIN starwars.appearances a ON c.idcharacters=a.id_characters
        JOIN starwars.media m ON m.idmedia=a.id_media
        WHERE LOWER(`source_name`) = '{}';'''.format(answer4a.lower()))
        fetchm = cur.fetchall()
        try:
            O = fetchm[0]['SDC']
        except TypeError:
            O = fetchm[0][0]
        print('The number of species which appear in {} is'.format(answer4a), O)
```

```python
def speciesasone(ex):
    answer4a = ((input('What character? ')))
    if ex == '1':
        cur.execute('''SELECT `names` AS 'nm' FROM starwars.characters c
        WHERE c.id_species IN 
        (SELECT id_species FROM starwars.characters 
        WHERE LOWER(`names`) = '{}');'''.format(answer4a.lower()))
        fetchm = cur.fetchall()
        try:
            species_species = [fm['nm'] for fm in fetchm]
        except TypeError:
            species_species = [fm[0] for fm in fetchm]
        print('Here is a list of all individuals of the same species as {}:'.format(answer4a))
        print(*species_species, sep = ', ')
    elif ex == '2':
        cur.execute('''SELECT COUNT(*) AS 'CS' FROM starwars.characters c
        WHERE c.id_species IN (SELECT id_species FROM starwars.characters WHERE LOWER(`names`) = '{}')
        GROUP BY id_species;'''.format(answer4a.lower()))
        fetchm = cur.fetchall()
        try:
            O = fetchm[0]['CS']
        except TypeError:
            O = fetchm[0][0]
        print('The number of individuals of the same species as {} is'.format(answer4a), O)
```
